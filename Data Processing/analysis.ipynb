```python
import numpy as np
import pandas as pd
import datetime
```

```python
history1 = pd.read_csv("Listening History Data - User 1.csv",na_values='null')
history2 = pd.read_csv("Listening History Data - User 2.csv",na_values='null')
history1['Id']='1'
history2['Id']='2'
```

```python
# concatenate two files
history = pd.concat([history1, history2], axis=0)
#history.to_csv("ListeningHistory.csv", index=False)
```

```python
print 'column:', history.shape[1]
print 'row:', history.shape[0]
history.describe()
```

### Time

```python
# Add date, hour and weekday from timestamp
history_time = history
history_time['date'] = history_time.apply(lambda x:
                                          datetime.datetime.fromtimestamp(x['timestamp']).strftime('%Y-%m-%d'),axis=1)
history_time['hour'] = history_time.apply(lambda x: 
                                          datetime.datetime.fromtimestamp(x['timestamp']).strftime('%H'),axis=1)
history_time['weekday'] = history_time.apply(lambda x:
                                          datetime.datetime.fromtimestamp(x['timestamp']).strftime('%a'),axis=1)
history_time.head()
```

```python
# When did the users like listening
pd.pivot_table(history_time,index=['hour'],values=["timestamp"],columns=["Id"],aggfunc=[np.size])
```

### Platform

```python
# all platforms used by two users
test = history_time.groupby(['Id','platform'])
test.size()
```

```python
test = history_time.groupby(['Id','weekday','platform'])
with pd.option_context('display.max_rows', 999, 'display.max_columns', 3):
    print test.size()
```

```python
# Split platform into device and system
history_device = history_time
history_device['device'] = history_time.apply(lambda x:x['platform'].replace('(','[').split('[')[1][:-1],axis=1)
history_device['system'] = history_time.apply(lambda x:x['platform'].replace('(','[').split('[')[0][:-1],axis=1)
history_device.drop('platform', axis=1, inplace=True)
history_device.head()
```

```python
# know the devices used on each day
test = history_device.groupby(['Id','date','device','system'])
with pd.option_context('display.max_rows', 999, 'display.max_columns', 3):
    print test.size()
```

```python
# sources used by different users
pd.pivot_table(history_device,index=['source'],values=["timestamp"],columns=["Id"],aggfunc=[np.size])
```

```python
pd.pivot_table(history_device,index=['Id'],values=['valence_score','danceability_score','familiarity_score',
                                                   'song_year','song_tempo','runnability','energy','popularity_normalized'],
               aggfunc=[np.mean])
```

```python
# almost no overlapping in songs listened
pd.pivot_table(history_device,index=['artist_id'],values=['timestamp'],columns=['Id'],
               aggfunc=[np.size])
```

### Normalize song_year

```python
history_std = history_device
history_std['up_to_date_score'] = (history_std.song_year - history_std.song_year.min()) / (history_std.song_year.max()-history_std.song_year.min())
history_std['tempo_score'] = (history_std.song_tempo - history_std.song_tempo.min()) / (history_std.song_tempo.max()-history_std.song_tempo.min())
```

```python
history_std.to_csv("ListeningHistory.csv", index=False)
```
